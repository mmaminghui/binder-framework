```python
NTA_grp=df.groupby(['NTA'])
NTA_grp['Company Name'].value_counts().head(50)
NTA_grp['Company Name'].nunique()
```

```python
NTA_grp['Company Name'].nunique()
NTA_grp['Total Savings'].mean()
NTA_grp['Job created'].sum()
NTA_grp_df=pd.concat([NTA_grp['Company Name'].nunique(),NTA_grp['Total Savings'].mean(),NTA_grp['Job created'].sum()], axis='columns', sort=False)
print(NTA_grp_df)
NTA_grp_df.rename(columns={'Company Name':'Business','Total Savings':'Average Total Savings','Job created':'Total Jobs'},inplace=True)
print(NTA_grp_df)

```
